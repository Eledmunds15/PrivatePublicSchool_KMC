K_Means Cluster Creation

```python
from sklearn.cluster import KMeans

KMM = KMeans(n_clusters=2, random_state=2312321)
KMM.fit(df.drop('Private', axis=1))
```

```python
KMM.labels_
```

Changing 'Private' Column to binary representation

```python
def clusterer(clusters):
    if clusters == 'Yes':
        return 1;
    elif clusters == 'No':
        return 0;
```

```python
df['Cluster'] = df['Private'].apply(clusterer)
df.head()
```

Checking Metrics for KMC Model

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(df['Cluster'], KMM.labels_))
print('\n')
print(confusion_matrix(df['Cluster'], KMM.labels_))
```
